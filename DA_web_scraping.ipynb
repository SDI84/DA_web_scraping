Создайте программу web-scraping. В качестве источника использовать https://quotes.toscrape.com/. Реализовать HTML разбор текста с выводом результата в виде DataFrame.

# Реализация:

```python
#Подключаем используемые библиотеки
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#выполним подключение к сайту https://quotes.toscrape.com/ ответ 200 указывает об успешном выполненом запросе
url = 'https://quotes.toscrape.com/'
html = requests.get(url)
html
```

```python
#извлечем данные из подключенного html с помощью BeautifulSoup
bs_text = BeautifulSoup(html.text, 'html.parser')
bs_text
```

```python
#запишем все части страницы сожержащие искомый текст в список list_text
list_text = bs_text.find_all('div',{'class':'quote'})
list_text[0]
```

```python
# создадим результирующий словарь со всеми Цитати | Авторами |  Списками тегов
result_text_author = {'Text':[],'Author':[],'Tags':[]}
for text_text in list_text:
    result_text_author['Text'].append(text_text.find('span',{'class':'text'}).get_text())
    result_text_author['Author'].append(text_text.find('small',{'class':'author'}).get_text())
    result_text_author['Tags'].append(','.join([l.get_text() for l in text_text.find_all('a',{'class':'tag'})]))
result_text_author
```

```python
#Выведем результат в DataFrame
result_data_frame = pd.DataFrame(data=result_text_author)
result_data_frame.head()
```
